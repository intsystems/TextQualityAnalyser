```python
from isanlp_rst.parser import Parser
import io, contextlib
import isanlp_rst

version = 'rstreebank'  # Choose from {'gumrrg', 'rstdt', 'rstreebank'}

parser = Parser(hf_model_name='tchewik/isanlp_rst_v3',
                hf_model_version=version,
                cuda_device=0) # Use -1 for CPU
```

```python
text1 = """Регулярные физические упражнения укрепляют иммунную систему. Это происходит потому, что умеренная нагрузка стимулирует выработку лейкоцитов.
В частности, уже 30 минут ходьбы в день повышают активность NK-клеток.Поэтому люди, занимающиеся спортом регулярно, болеют простудой в среднем на 40% реже."""
```

```python
res = parser(text1)
res['rst'][0].to_rs3('text1.rs3')

buf = io.StringIO()
with contextlib.redirect_stdout(buf):
    isanlp_rst.render("text1.rs3", colab=True)
```

```python
text2 = "Спорт укрепляет сердечно-сосудистую систему. Спорт снижает уровень стресса. Спорт помогает поддерживать нормальный вес. Спорт улучшает качество сна"
```

```python
res = parser(text2)
res['rst'][0].to_rs3('text2.rs3')

# --- workaround для визуализации ---
with open('text2.rs3', 'r', encoding='utf-8') as f:
    rs3 = f.read()

if '<rel ' in rs3 and 'type="rst"' not in rs3:
    rs3 = rs3.replace(
        '</relations>',
        '\t\t\t<rel name="elaboration" type="rst" />\n\t\t</relations>'
    )
if 'type="multinuc"' not in rs3:
    rs3 = rs3.replace(
        '</relations>',
        '\t\t\t<rel name="joint" type="multinuc" />\n\t\t</relations>'
    )

with open('text2.rs3', 'w', encoding='utf-8') as f:
    f.write(rs3)
# ---------------------------------------------------------------------


buf = io.StringIO()
with contextlib.redirect_stdout(buf):
  isanlp_rst.render("text2.rs3", colab=True)
```
